```python
import matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
X = [1,   1.5, 2,   3, 5, 6]
Y = [1.8, 1,   3.6, 2, 1, 3]
plt.plot(X, Y)
plt.show()
```

```python
plt.scatter(X,Y)  # рисуем точки
plt.title("Просто точки")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.show()
```

```python
plt.scatter(X, Y)
plt.xlim(0,4)
plt.ylim(0,2)
plt.show()
```

```python
plt.scatter(X, Y, s=[20, 60, 30, 100, 60, 10], c='plum', marker='^')
plt.show()
```

```python
size = [60, 70, 80, 90, 100, 130]
colors = ['red', 'green', 'blue', 'black', 'yellow', 'brown']
plt.scatter(X, Y, s=size, c=colors, marker='^')
plt.show()
```

```python
markers = ['o', '*', '8', 'D', '^', 'P']

for x, y, m in zip(X, Y, markers):
    plt.scatter([x],[y], marker=m, s=100)
plt.show()
```

```python
dots = ['Первая', "Вторая", "Третья", "Четвертая", "Пятая", "Шестая"]
from matplotlib import style
style.use('ggplot')
for x, y, m, d in zip(X, Y, markers, dots):
    plt.scatter(x, y, marker=m, s=100)
    plt.text(x+0.1, y, d) # +0.1 - это чтобы текст не наползал на маркер, а отрисовывался чуть выше и правее 
plt.show()
```

```python

plt.plot(range(10), linestyle='--', marker='D', c='xkcd:blue', markevery=[4, 7,8])
plt.show()
```

```python
plt.plot(range(10), '-.cD', markevery=[2, 3,6])
```

```python
A = [2, 4, 6, 6, 10, 8, 4, 10, 3, 2]
B = [3, 6, 9, 9, 9, 10, 0, 3, 5, 7]
from matplotlib import style
style.use('ggplot')
plt.plot(range(10), A, c='b', label="smops")
plt.plot(range(10), B, c='y', label="pops")
plt.title("Котики и пёсики")
plt.xlabel("Какой-то номер")
plt.ylabel("Какая-то частота")
plt.legend()

plt.show()
```

```python

A = [2, 4, 6, 6, 10, 8, 4, 10, 3, 2]

plt.scatter(range(10), A, c='b', marker="8", s=100)
plt.plot(range(10), [5] * 10, c='y')
plt.title("Пупочки")
plt.xlabel("Номер текста")
plt.ylabel("Абсолютная частота")

plt.show()
```

```python
# сохранять график надо перед тем, как его выводить, 
# т.к. после команды plt.show() график очищается

plt.plot(range(10), '--gD', markevery=[5, 6,7])
plt.savefig('my_first_plot.pdf')
plt.show()
```

```python
plt.plot(X, Y, 'r')
plt.savefig('my_first_plot.png', dpi=300)
plt.show()
```
