2. Очистите текст от пунктуации, приведите его к нижнему регистру, лемматизируйте и запишите в новый файл. Проверьте, выполняется ли здесь закон Ципфа. Для этого постройте линейную диаграмму, где по оси Х будет ранг (порядковый номер) слова в отсортированном по убыванию частотном словаре, а по оси Y - относительная частота (абсолютная частота, разделенная на количество слов в тексте). Т.е. у самого частотного слова будет ранг 1, у второго по частоте -- 2 и т.д.

```python
#Очищаем текст от пуктуации
words = []

for word in book.split():
    words.append(word.strip("!\"#$%&'()*+,./:;<=>?@[\]^_`{|}~„“«»†*—/\-—"))
    
print(words)
```

```python
# Приводим текст к нижнему регистру
book2 =' '.join(words) 
book3 = book2.lower()
print(book3)
```

```python
#Лемматизируем текст 
from nltk import WordNetLemmatizer

wnl = WordNetLemmatizer()
wnl.lemmatize(book3)
#и записываем в новый файл
with open('Harry_lem.txt', 'w', encoding='utf-8') as f:
    f.write(book3)
# Узнаем общее количество слов, чтобы проверить закон Ципфа
len(book3)
```

```python
# Cоздаем отсортированный по частотности список слов
from collections import Counter
counts = Counter(book3.split())

print(counts)
```

```python
#Проверим закон Ципфа, построив диаграмму.
#Для этого возьмем первые 6 значений и рассчитаем их относительную частоту с помощью ipm
a = 3603/415773*1000000
b = 1870/415773*1000000
c = 1845/415773*1000000
d = 1667/415773*1000000
f = 1497/415773*1000000
h = 1255/415773*1000000
print (a,b,c,d,f,h)
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style  # добавляем стили
style.use('ggplot')

# строим диаграмму
X = [1, 2, 3, 4, 5, 6]
Y = [8665.7, 4497.6, 4437.5, 4009.3, 3600.5, 3018.4]
plt.plot(X, Y)
plt.title("Закон ципфа")
plt.xlabel("Ранг (порядковый номер)")
plt.ylabel("Относительная частота")

plt.show()
```

3. Очистите текст от стоп-слов. Постройте barplot для 30 самых частотных слов. Эти слова должны быть по оси Х в виде подписей к делениям. Значениями по этой оси будет ряд чисел от 0 до 30.

```python
#Очищаем текст от стоп-слов
from nltk.corpus import stopwords
#загружаем english стоп-слова
stopwords.fileids()
stop = set(stopwords.words('english'))
print (stop)
```

```python
#Очищаем от стоп-слов

from nltk.tokenize import word_tokenize
print(word_tokenize(book3))
from nltk.tokenize import sent_tokenize

sent_tokenize(book3)
```

```python
sw = stopwords.words('english')

filtered = [w for w in word_tokenize(book3) if w not in sw]
print(filtered)
```

```python
# Снова создаем отсортированный по частотности список слов
filtered2 =' '.join(filtered) 
from collections import Counter
counts2 = Counter(filtered2.split())

print(counts2)
```
